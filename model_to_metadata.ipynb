```python
from tflite_support.metadata_writers import object_detector
from tflite_support.metadata_writers import writer_utils
```

```python
import tempfile
import os
from google.colab import files

# Create ObjectDetectorWriter
ObjectDetectorWriter = object_detector.MetadataWriter

# Upload model TFLite Anda ke Colab
print("Silakan upload file model TFLite Anda:")
uploaded = files.upload()

# Ambil nama file yang diupload
model_filename = list(uploaded.keys())[0]
_MODEL_PATH = model_filename
_SAVE_TO_PATH = f"{model_filename.split('.')[0]}_with_metadata.tflite"

print(f"Model yang diupload: {_MODEL_PATH}")
print(f"Output akan disimpan sebagai: {_SAVE_TO_PATH}")

# Buat file label berdasarkan metadata.yaml
labels = [
    "matang",
    "sangat-busuk",
    "sedikit-busuk"
]

# Buat temporary label file
with tempfile.NamedTemporaryFile(mode='w', suffix='.txt', delete=False) as f:
    for label in labels:
        f.write(label + '\n')
    _LABEL_FILE = f.name

# Normalization parameters untuk YOLOv8
# YOLOv8 biasanya menggunakan normalisasi 0-1, jadi:
# pixel_value = (input - 0) / 255.0
_INPUT_NORM_MEAN = 0.0
_INPUT_NORM_STD = 255.0

# Atau jika menggunakan standardisasi ImageNet (lebih umum untuk YOLOv8):
# _INPUT_NORM_MEAN = [0.485 * 255, 0.456 * 255, 0.406 * 255]  # RGB channels
# _INPUT_NORM_STD = [0.229 * 255, 0.224 * 255, 0.225 * 255]   # RGB channels

try:
    # Create the metadata writer
    writer = ObjectDetectorWriter.create_for_inference(
        writer_utils.load_file(_MODEL_PATH),
        [_INPUT_NORM_MEAN],
        [_INPUT_NORM_STD],
        [_LABEL_FILE]
    )

    # Verify the metadata generated by metadata writer
    print("=== Generated Metadata JSON ===")
    print(writer.get_metadata_json())

    # Populate the metadata into the model
    writer_utils.save_file(writer.populate(), _SAVE_TO_PATH)

    print(f"\n=== Success! ===")
    print(f"Model dengan metadata telah disimpan ke: {_SAVE_TO_PATH}")

    # Download file hasil
    print("Download file hasil:")
    files.download(_SAVE_TO_PATH)

    # Model info berdasarkan tensor yang Anda berikan:
    print(f"\n=== Model Information ===")
    print(f"Input tensor name: images")
    print(f"Input shape: [1, 640, 640, 3]")
    print(f"Input dtype: float32")
    print(f"Output tensor name: Identity")
    print(f"Output shape: [1, 7, 8400]")
    print(f"Output dtype: float32")
    print(f"Classes: {len(labels)} ({', '.join(labels)})")

finally:
    # Clean up temporary label file
    if os.path.exists(_LABEL_FILE):
        os.unlink(_LABEL_FILE)

print("\n=== Catatan Penggunaan di Google Colab ===")
print("1. Jalankan kode ini di Google Colab")
print("2. Upload file model TFLite Anda ketika diminta")
print("3. File hasil akan otomatis didownload setelah selesai")
print("4. Output tensor shape [1, 7, 8400] menunjukkan:")
print("   - 1: batch size")
print("   - 7: [x, y, w, h, confidence, class_prob1, class_prob2, class_prob3]")
print("   - 8400: jumlah anchor boxes (640x640 input dengan berbagai scale)")
print("5. Untuk inference, Anda perlu post-processing untuk NMS dan threshold")
```
